```python
import pandas as pd

sales = pd.read_csv("stadiagear_clean.csv")
customer = pd.read_csv("Customer_Dimension.csv")
```

```python
customer.columns
```

```python
sales_customer = pd.merge(
    sales,
    customer,
    on='CustomerID',
    how='left'
)
```

```python
sales_customer.head()
```

```python
segment_revenue = sales_customer.groupby(
    'CustomerSegment'
)['Revenue'].sum().reset_index()
```

```python
segment_revenue
```

```python
customer.columns
```

```python
segment_revenue
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))

sns.barplot(
    data=segment_revenue,
    x='CustomerSegment',
    y='Revenue'
)

plt.title('Revenue by Customer Segment')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=segment_revenue,
    x='CustomerSegment',
    y='Revenue'
)

plt.title('Revenue by Customer Segment')

plt.savefig('Revenue_By_Segment.png')

plt.show()
```
