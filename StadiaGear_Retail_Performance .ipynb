```python
num_cols = [
    'Quantity',
    'UnitPrice',
    'Revenue',
    'InventoryUnits',
    'UnitsSold',
    'SellThroughRatePct',
    'CustomerLifetimeValue'
]
```

```python
corr = df[num_cols].corr()

corr
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
top_cat = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)

top_cat.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Revenue by Category")
plt.show()
```

```python
top_team = df.groupby('Team_Affiliation')['Revenue'].sum().sort_values(ascending=False)

top_team.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Revenue by Team")
plt.show()
```
