```python
import pandas as pd

product = pd.read_csv("Product_Dimension.csv")
```

```python
product.head()
product.info()
product.describe()
```

```python
product.isnull().sum()
```

```python
product.duplicated().sum()
```

```python
product.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(data=product, x='Category')
plt.title("Category Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(product['UnitPrice'], bins=20, kde=True)
plt.title("Unit Price Distribution")
plt.show()
```

```python
product['Category'].value_counts()
```

```python
product['Category'].value_counts().plot(kind='bar')
plt.title("Products per Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=product,
    x='Category',
    y='UnitPrice'
)

plt.title("Average Unit Price by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=product['UnitPrice'])
plt.title("Unit Price Distribution")
plt.show()
```
