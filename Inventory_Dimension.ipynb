```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

inventory = pd.read_csv("Inventory_Dimension.csv")
```

```python
inventory.head()
```

```python
inventory.shape
```

```python
inventory.info()
```

```python
inventory.describe()
```

```python
inventory.isnull().sum()
```

```python
inventory.fillna(0, inplace=True)
```

```python
inventory.duplicated().sum()
```

```python
inventory.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    inventory['InventoryUnits'],
    bins=20,
    kde=True
)

plt.title("Inventory Units Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    inventory['UnitsSold'],
    bins=20,
    kde=True
)

plt.title("Units Sold Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    inventory['SellThroughRatePct'],
    bins=20,
    kde=True
)

plt.title("Sell Through Rate Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=inventory,
    x='InventoryUnits',
    y='UnitsSold'
)

plt.title("Inventory vs Units Sold")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=inventory,
    x='InventoryUnits',
    y='SellThroughRatePct'
)

plt.title("Inventory vs Sell Through Rate")
plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=inventory['InventoryUnits']
)

plt.title("Inventory Units Outliers")
plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=inventory['UnitsSold']
)

plt.title("Units Sold Outliers")
plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=inventory['SellThroughRatePct']
)

plt.title("Sell Through Rate Outliers")
plt.show()
```
