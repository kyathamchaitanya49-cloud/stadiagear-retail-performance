```python
customer.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    customer['CustomerLifetimeValue'],
    bins=20,
    kde=True
)

plt.title("Customer Lifetime Value Distribution")
plt.show()
```

```python
customer['CustomerSegment'].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=customer,
    x='CustomerSegment'
)

plt.title("Customer Segments")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=customer,
    x='Team_Affiliation'
)

plt.title("Team Affiliation")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=customer,
    x='CustomerSegment',
    y='CustomerLifetimeValue'
)

plt.title("Average CLV by Customer Segment")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=customer,
    x='Team_Affiliation',
    y='CustomerLifetimeValue'
)

plt.title("Average CLV by Team")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=customer['CustomerLifetimeValue']
)

plt.title("Customer Lifetime Value Distribution")
plt.show()
```
