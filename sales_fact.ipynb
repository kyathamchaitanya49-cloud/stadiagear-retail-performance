```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Sales_Fact.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Quantity'], bins=20)
plt.title("Quantity Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Revenue'], bins=20)
plt.title("Revenue Distribution")
plt.show()
```

```python
sns.countplot(data=df,x='Region')
plt.title("Region Count")
plt.show()
```

```python
df['Region'].value_counts()
```

```python
df['Region'].value_counts().plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

```python
df['ProductID'].value_counts().head(10)
```

```python
df['ProductID'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Products")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df,x='Quantity',y='Revenue')
plt.title("Revenue vs Quantity")
plt.show()
```

```python
sns.barplot(data=df,x='Region',y='Revenue')
plt.title("Revenue by Region")
plt.show()
```

```python
sns.histplot(df['Revenue'], kde=True)
plt.title("Revenue Distribution")
plt.show()
```

```python
sns.histplot(df['Quantity'], kde=True)
plt.title("Quantity Distribution")
plt.show()
```

```python
sns.boxplot(x=df['Revenue'])
plt.title("Revenue Outliers")
plt.show()
```

```python
sns.boxplot(x=df['Quantity'])
plt.title("Quantity Outliers")
plt.show()
```

```python
corr = df[['TransactionID',
           'CustomerID',
           'Quantity',
           'Revenue']].corr()

corr
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
corr = df[['Quantity','Revenue']].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Quantity vs Revenue Correlation")
plt.show()
```
